```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dpvb = [1.01430034, 0.95240449 , 0.70544731,0.69566494, 0.68388992, 0.44569686, 0.38633611]

xmvb = [0.03769585, 0.03563491, 0.02587526, 0.02536703, 0.02498515, 0.01560112,0.01365898]

fig, ax1 = plt.subplots(figsize=(8, 6))

for spine in plt.gca().spines.values():
    spine.set_linewidth(2)

xmvb = [x / max(xmvb) for x in xmvb]
dpvb = [x / max(dpvb) for x in dpvb]

# 绘制剩余点的折线图
ax1.plot(np.arange(len(dpvb)), dpvb, marker='*', markersize=10, linestyle='--', color='blue', label='DPVB')

ax1.set_xlabel('VB Structures', fontsize=22)
ax1.set_ylabel('Weight', color='blue', fontsize=30)
ax1.tick_params('y', colors='blue')
ax1.legend(loc='upper right')

# ax2 = ax1.twinx()
ax1.plot(np.arange(len(xmvb)), xmvb, marker='o', markersize=10, linestyle='--', color='red', label='XMVB')
ax1.set_ylabel('Relative Weight', color='blue', fontsize=22)
ax1.tick_params('y', colors='black')
ax1.legend(loc='upper right',bbox_to_anchor=(1,0.98), fontsize=22, frameon=False)

# 设置 x 和 y 轴刻度线字体大小
ax1.tick_params(axis='x', labelsize=15)  # x轴刻度字体大小
ax1.tick_params(axis='y', labelsize=15)  # y轴刻度字体大小

# 调整布局并显示图表
plt.tight_layout()
plt.savefig('Rank_line.png', dpi=600)
plt.show()
```

```python
dpvb = [1.01430034, 0.95240449 , 0.70544731,0.69566494, 0.68388992, 0.44569686, 0.38633611]

xmvb = [0.03769585, 0.03563491, 0.02587526, 0.02536703, 0.02498515, 0.01560112,0.01365898]

fig, ax1 = plt.subplots(figsize=(8, 6))

for spine in plt.gca().spines.values():
    spine.set_linewidth(2)

xmvb = [x / max(xmvb) for x in xmvb]
dpvb = [x / max(dpvb) for x in dpvb]

# 绘制剩余点的折线图
ax1.scatter(np.arange(len(dpvb)), dpvb, marker='*',  linestyle='--', color='blue', label='DLVB', s=150)

ax1.set_xlabel('VB Structure', fontsize=22)
ax1.set_ylabel('Weight', color='blue', fontsize=30)
ax1.tick_params('y', colors='blue')
ax1.legend(loc='upper right')

# ax2 = ax1.twinx()
ax1.scatter(np.arange(len(xmvb)), xmvb, marker='^', linestyle='--', color='red', label='Exact', s=150)
ax1.set_ylabel('Relative Weight', color='blue', fontsize=22)
ax1.tick_params('y', colors='black')
ax1.legend(loc='upper right',bbox_to_anchor=(1,0.98), fontsize=22)

# 设置 x 和 y 轴刻度线字体大小
ax1.tick_params(axis='x', labelsize=15)  # x轴刻度字体大小
ax1.tick_params(axis='y', labelsize=15)  # y轴刻度字体大小

# 调整布局并显示图表
plt.tight_layout()
plt.savefig('Rank_scatter.png', dpi=600)
plt.show()
```
