```python
import matplotlib.pyplot as plt
import numpy as np
import re
import json

def read_log(log_file, target="auc"):
    losses = []
    with open(log_file, 'r') as file:
        for line in file:
            record = json.loads(line)  # 将每一行解析为 JSON 对象
            loss = record[target]
            losses.append(loss)

    return losses
print(plt.style.available)
```

plt the binary classification results of the training set

```python

bi_val_file = '/home/xiatao/vbnet/GraphGPS-main/results/Final_binary/0/val/stats.json'
bi_teat_file = '/home/xiatao/vbnet/GraphGPS-main/results/Final_binary/0/test/stats.json'
bi_train_file = '/home/xiatao/vbnet/GraphGPS-main/results/Final_binary/0/train/stats.json'
train_auc = read_log(bi_train_file)
val_auc = read_log(bi_val_file)
test_auc = read_log(bi_teat_file)
epochs = range(len(train_auc))

plt.style.use('seaborn-v0_8-ticks')  # 或其他可用样式
# 绘制曲线
plt.figure(figsize=(8, 6))
plt.plot(epochs, train_auc, label='Train Accuracy', color='blue', linewidth=2, linestyle='-')
plt.plot(epochs, val_auc, label='Validation Accuracy', color='green', linewidth=2, linestyle='--')
plt.plot(epochs, test_auc, label='Test Accuracy', color='red', linewidth=2, linestyle='-.')


# 添加网格和标题
# plt.grid(alpha=0.5, linestyle='--')
plt.xlabel('Epoch', fontsize=20)
plt.ylabel('Accuracy', fontsize=20)
plt.title('Training, Validation, and Test Loss (Accuracy)', fontsize=20)
plt.grid(alpha=0.3, linestyle='--')
plt.xticks(fontsize=16)  # 设置 x 轴刻度字体大小
plt.yticks(fontsize=16)  # 设置 y 轴刻度字体大小
plt.legend(fontsize=20, loc='lower right')
ax = plt.gca()  # 获取当前坐标轴
ax.spines['top'].set_linewidth(2)  # 设置上边框宽度
ax.spines['bottom'].set_linewidth(2)  # 设置下边框宽度
ax.spines['left'].set_linewidth(2)  # 设置左边框宽度
ax.spines['right'].set_linewidth(2)  # 设置右边框宽度
plt.tight_layout()
plt.savefig("model_accuracy.png", dpi=600)
plt.show()

```

```python
bi_val_file = '/home/xiatao/vbnet/GraphGPS-main/results/Final_rel/0/val/stats.json'
bi_test_file = '/home/xiatao/vbnet/GraphGPS-main/results/Final_rel/0/test/stats.json'
bi_train_file = '/home/xiatao/vbnet/GraphGPS-main/results/Final_rel/0/train/stats.json'
train_auc = read_log(bi_train_file, target='loss')
val_auc = read_log(bi_val_file, target='loss')
test_auc = read_log(bi_test_file, target='loss')
epochs = range(len(train_auc))

plt.figure(figsize=(8, 6))
plt.plot(epochs, train_auc, label='Train MAE', color='blue', linewidth=2, linestyle='-')
plt.plot(epochs, val_auc, label='Validation MAE', color='green', linewidth=2, linestyle='--')
plt.plot(epochs, test_auc, label='Test MAE', color='red', linewidth=2, linestyle='-.')

plt.xticks(fontsize=16)  # 设置 x 轴刻度字体大小
plt.yticks(fontsize=16)  # 设置 y 轴刻度字体大小
plt.legend(fontsize=20, loc='upper right')
plt.xlabel('Epoch', fontsize=20)
plt.ylabel('Loss (MAE)', fontsize=20)
plt.title('Training, Validation, and Test Loss (MAE)', fontsize=20)
plt.grid(alpha=0.3, linestyle='--')
ax = plt.gca()  # 获取当前坐标轴
ax.spines['top'].set_linewidth(2)  # 设置上边框宽度
ax.spines['bottom'].set_linewidth(2)  # 设置下边框宽度
ax.spines['left'].set_linewidth(2)  # 设置左边框宽度
ax.spines['right'].set_linewidth(2)  # 设置右边框宽度
plt.tight_layout()
plt.savefig('model_mae.png', dpi=600)
plt.show()
```
