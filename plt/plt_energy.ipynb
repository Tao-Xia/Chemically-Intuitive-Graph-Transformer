```python
import matplotlib.pyplot as plt

# 
subspaces = ['cov','0-1','DLVB', '0-2', '0-3',  'Full']

# 
nstr_values = [14, 294, 476, 1134, 1694,1764]

# 
E_values = [130.9551, 24.35179, 0.4952123, 1.836477, 0.0283177,  0.0]

# 
time = [200.05, 243.85, 275.63, 533.80, 873.49,  880.06]  # 

# 
fig, ax1 = plt.subplots()

# 
ax1.plot(subspaces, E_values, marker='o', linestyle='--', color='blue', label='Energy Error (Kcal/mol)')
ax1.set_xlabel('Subspace')
ax1.set_ylabel('Energy Error', color='blue')
ax1.tick_params('y', colors='blue')

#  DLVB  marker
DLVB_index = subspaces.index('DLVB')
ax1.plot(subspaces[DLVB_index], E_values[DLVB_index], marker='*', color='blue', markersize=10
         )

# x
ax2 = ax1.twinx()

# 
ax2.plot(subspaces, time, marker='o', linestyle='--', color='red', label='Calculation Time (sec)')
ax2.set_ylabel('Calculation Time', color='red')
ax2.tick_params('y', colors='red')
ax2.plot(subspaces[DLVB_index], time[DLVB_index], marker='*', color='red', markersize=10
         )
# 
for i, subspace in enumerate(subspaces):
    ax1.text(i, E_values[i], str(nstr_values[i]), ha='center', va='bottom')

# 
ax1.legend(loc='center right', bbox_to_anchor=(0.53, 0.9))
ax2.legend(loc='center right', bbox_to_anchor=(0.52, 0.8))

# 
plt.tight_layout()
# plt.savefig('Energy_compare_with_time.png', dpi=600)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 
subspaces = ['cov', '0-1', '0-2', '0-3', 'DLVB', 'Full']

# 
nstr_values = [14, 294, 1134, 1694, 476, 1764]

# 
E_values = [130.9551, 24.35179, 1.836477, 0.0283177, 0.4952123, 0.0]

# 
time = [200.05, 243.85, 533.80, 873.49, 275.63, 880.06]  # 

# 
fig, ax1 = plt.subplots()

# 
bar_width = 0.6
x_positions = np.arange(len(subspaces))
ax1.bar(x_positions, time, color='lightsalmon', alpha=0.4, width=bar_width, label='Calculation Time (sec)')

# x
ax1.set_xticks(x_positions)
ax1.set_xticklabels(subspaces)

# y
ax1.set_xlabel('Subspace')
ax1.set_ylabel('Calculation Time', color='red')
ax1.tick_params('y', colors='red')

# y
ax2 = ax1.twinx()

# 
ax2.plot(subspaces, E_values, marker='o', linestyle='--', color='blue', label='Energy Error (Kcal/mol)')
ax2.set_ylabel('Energy Error', color='blue')
ax2.tick_params('y', colors='blue')

# 
for i, subspace in enumerate(subspaces):
    ax2.text(i, E_values[i], str(nstr_values[i]), ha='center', va='bottom')

# 
ax1.legend(loc='upper left', bbox_to_anchor=(0.05, 0.9))
ax2.legend(loc='upper left', bbox_to_anchor=(0.05, 0.8))

# 
plt.tight_layout()
# plt.savefig('Energy_compare_with_time_bar.png', dpi=600)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 
subspaces = ['cov', '0-1', 'DLVB', '0-2', '0-3', 'Full']

# 
subspace_legend = [
    'cov : Covalent', 
    '0-1 : Covalent + 1st Ionic', 
    'DLVB',  # 
    '0-2 : Covalent + 1st-2nd Ionic', 
    '0-3 : Covalent + 1st-3rd Ionic', 
    'Full : Full Structure'
]

# 
nstr_values = [14, 294, 441, 1134, 1694, 1764]

# 
E_values = [114.381, 19.459, 1.011, 1.220, 0.017, 0.000 ]

# 
fig, ax1 = plt.subplots(figsize=(8, 6))
for spine in plt.gca().spines.values():
    spine.set_linewidth(3)

# DLVB
nstr_values_without_DLVB = [n for i, n in enumerate(nstr_values) if subspaces[i] != 'DLVB']
E_values_without_DLVB = [E for i, E in enumerate(E_values) if subspaces[i] != 'DLVB']

# 
ax1.plot(nstr_values_without_DLVB, E_values_without_DLVB, marker='o', linestyle='--', color='blue')

# DLVB
DLVB_index = subspaces.index('DLVB')
ax1.scatter(nstr_values[DLVB_index], E_values[DLVB_index], color='blue', marker='*', s=100, label='DLVB')

# 
ax1.set_xlabel('Number of VB Structures', fontsize=20)
ax1.set_ylabel('Energy Error (Kcal/mol)', color='blue', fontsize=20)
ax1.tick_params('y', colors='black')

# 
for i, subspace in enumerate(subspaces):
    if subspace != 'DLVB':  # DLVB
        ax1.scatter(nstr_values[i], E_values[i], color='blue', label=subspace_legend[i])

for i, subspace in enumerate(subspaces):
    if subspace != 'DLVB':  # DLVB
        if i == 0:
            ax1.text(nstr_values[i] + 100, E_values[i] - 5, subspace, color='black', fontsize=18, ha='center', va='bottom')
        elif i > 0 and i <= 3:
            ax1.text(nstr_values[i] + 60, E_values[i], subspace, color='black', fontsize=18, ha='center', va='bottom')
        elif i == 4:
            ax1.text(nstr_values[i] - 40, E_values[i] + 2, subspace, color='black', fontsize=18, ha='center', va='bottom')
        elif i == 5:
            ax1.text(nstr_values[i] + 12, E_values[i] + 2, subspace, color='black', fontsize=18, ha='center', va='bottom')
    else:
        ax1.text(nstr_values[i], E_values[i] + 2, subspace, color='black', fontsize=18, ha='center', va='bottom')

# 
handles, labels = ax1.get_legend_handles_labels()
by_label = dict(zip(labels, handles))
ax1.legend(by_label.values(), by_label.keys(), loc='upper right', fontsize=18, frameon=False)

ax1.tick_params(axis='x', labelsize=15)  # x
ax1.tick_params(axis='y', labelsize=15)  # y
# 
plt.tight_layout()
plt.savefig('Energy-2782798.png', dpi=600)
plt.show()
```
